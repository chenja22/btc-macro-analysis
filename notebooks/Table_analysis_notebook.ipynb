```python
import pandas as pd
from IPython.display import display

def show_table(file, name):
    data = pd.read_csv(file, parse_dates=['date'])
    stats = data[['log_return', 'sp500_return', 'vix_change', 'yield_change']].describe().T
    print(f"\nStats for {name}:")
    display(stats)
    stats.to_csv(f"../tables/{name}_table.csv")
    return stats

election = show_table("../data/btc_all_merged_clean.csv", "btc_election")
ratecut = show_table("../data/btc_all_ratecut_merged_clean.csv", "btc_ratecut")
```
